```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.collections as collections
```

Load results

```python
data = pd.read_csv('./extraction/XY10/clean_tracking_data.csv') # field of view 0
columns = data.columns.tolist()
data = data.reindex(columns=['particle', 'frame', 'nucleus', 'front', 'rear', 'fov', 'segment', 'state'])
```

```python
data.frame = data.frame*120/3600 # rescale frame to hours
```

Show traces of cell position

```python
p = 28 # particle 0
data_p = data.loc[data.particle == p]
fig, ax = plt.subplots(1, 1, figsize=(8,6))
ax.plot(data_p.frame, data_p.nucleus, label='nuclei', color='black')
ax.plot(data_p.frame, data_p.front, label='front', color='red')
ax.plot(data_p.frame, data_p.rear, label='rear', color='blue')
ax.legend()

plt.title('cell trajectories')
```
